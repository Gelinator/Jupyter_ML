```python
#Reference
#http://nbviewer.ipython.org/github/savarin/pyconuk-introtutorial/blob/master/notebooks/Section%201-0%20-%20First%20Cut.ipynb
import pandas as pd
import numpy as np

data = pd.read_csv("/Users/davegelinas/GitHub/LearningMachines/Prudential Challenge/train.csv",delimiter=',')
```

```python
data['Product_Info_2'] = data['Product_Info_2'].map({'D3':0, 'A1':1, 'E1':2, 'D4':3, 'D2':4, 'A8':5, 'A2':6, 'D1':7, 'A7':8, 'A6':9, 'A3':10,'A5':11, 'C4':12, 'C1':13, 'B2':14, 'C3':15, 'C2':16, 'A4':17, 'B1':18}).astype(int)
```

```python
data.info()
```

```python
#Reference

#Reshape of numpy array
#http://stackoverflow.com/questions/21938932/how-do-i-convert-a-numpy-array-into-a-pandas-dataframe

#Imputer
#http://stackoverflow.com/questions/11441751/how-to-get-svms-to-play-nicely-with-missing-data-in-scikit-learn

from sklearn.preprocessing import Imputer
imp = Imputer(missing_values='NaN', strategy='median', axis=0, verbose=0, copy=True)
imp.fit(data)
data = imp.transform(data)
data = data.reshape((-1,128))
data = pd.DataFrame(data)
```

```python
#Reference
#subset dataframe
#http://www.datacarpentry.org/python-ecology/03-data-types-and-format

# create a Python list of feature names
# use the list to select a subset of the original DataFrame
X = data.iloc[0:,0:127]

# print the first 5 rows
X.head()
```

```python
# select a Series from the DataFrame
y = data.iloc[:,127:]

# equivalent command that works if there are no spaces in the column name
#y = data.Response

# print the first 5 values
y.head()
```

```python
#Making sure data and target are the same length
print(type(X))
print(X.shape)

print(type(y))
print(y.shape)
```

```python
#Creating cross validation samples
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#Model fit and metrics
#https://github.com/justmarkham/scikit-learn-videos/blob/master/05_model_evaluation.ipynb

from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
boost = GradientBoostingClassifier()
boost.fit(X, y)
y_pred = boost.predict(X)
print(metrics.accuracy_score(y, y_pred))
```
