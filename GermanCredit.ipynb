```python
#Reference
#http://nbviewer.ipython.org/github/savarin/pyconuk-introtutorial/blob/master/notebooks/Section%201-0%20-%20First%20Cut.ipynb
#Data from the UCI Machine Learning Repository : http://archive.ics.uci.edu/ml/datasets/Statlog+%28German+Credit+Data%29
import pandas as pd
import numpy as np

data = pd.read_csv("~/GitHub/LearningMachines/GermanCredit.csv",delimiter=',')
```

```python
#target = iris[1,4]
#data = iris
#data.columns = ["Sepal.Length","Sepal.Width","Petal.Length","Petal.Width","Species"]

data['Class'] = data['Class'].map({'Good': 0, 'Bad': 1}).astype(int)
data.head()
```

```python
data = data.iloc[:,2:]
data.head()
```

```python
# create a Python list of feature names
#feature_cols = ["Sepal.Length","Sepal.Width","Petal.Length","Petal.Width"]

#http://stackoverflow.com/questions/14940743/selecting-excluding-sets-of-columns-in-pandas

# use the list to select a subset of the original DataFrame
X = data.drop(['Class'], axis=1)

# equivalent command to do this in one line
#X = data[["Sepal.Length","Sepal.Width","Petal.Length","Petal.Width"]]

# print the first 5 rows
X.head()
```

```python
# select a Series from the DataFrame
y = data['Class']

# equivalent command that works if there are no spaces in the column name
#y = iris.Species

# print the first 5 values
y.head()
```

```python
#Making sure data and target are the same length
print(type(X))
print(X.shape)

print(type(y))
print(y.shape)
```

```python
#Creating cross validation samples
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1,test_size=0.2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
boost = GradientBoostingClassifier(n_estimators=160)
boost.fit(X_train, y_train)
y_pred = boost.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier(n_neighbors=500)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
rf = RandomForestClassifier(n_estimators=1000)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred))
```
